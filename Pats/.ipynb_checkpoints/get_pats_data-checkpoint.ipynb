```python
import nflgame
import re
import pandas as pd
import sys
```

```python
# Create lists for regular and postseason weeks and all years from 2009
weeks=list(range(1,18))
print("weeks: " + str(weeks))
weeks_playoff=list(range(1,5))
print("weeks_playoff: " + str(weeks_playoff))
years=list(range(2009,2020))
print("years: "+ str(years))
```

```python
#create a function for creating list from a game object
def create_list_game_plays (game):
    #Get all the plays from all drives in one game
    myplays=game.drives.plays()            
    
    #create list for get all plays in a game   
    list_game_plays= []
    
    #iterate on myplays        
    for mp in myplays:        
        list_play=[]        
        if "PENALTY" in str(mp):                        
            year= str(game.schedule['year'])

            #create regex
            
            re_team_poss= re.search(r"^\((\D*),",str(mp))
            if re_team_poss:
                team_possesion=re_team_poss.group(1)
            else:
                re_team_poss= None
            
            re_field_position= re.search(r"([A-Z]* \d*), Q",str(mp))
            if re_field_position:
                field_position=re_field_position.group(1)
                #print(field_position)
                #regex for team and yards
                re_team_field=re.search(r"(\w+)* (\d+)",field_position)
                re_team_yards=re.search(r"(\w+)* (\d+)",field_position)
                #assign variables
                if re_team_field:
                    field_team=re_team_field.group(1)
                else:
                    field_team = None                
                field_yards=re_team_yards.group(2)
                field_yards.strip()
                field_yards = int(field_yards)
                
                
            else:
                re_field_position= None
    
            

            re_quarter=re.search(r"(Q\d)",str(mp))
            if re_quarter:
                quarter=re_quarter.group()
            else:
                re_quarter= None

            re_time=re.search(r"(\d*:\d+)",str(mp))
            if re_time :
                time=re_time.group(1)
                #fill minutes when we hve only seconds i.e (:10)
                re_minutes=re.search(r"(\d*):",time)
                no_minutes=re_minutes.group()
                if no_minutes == ':':
                    time='00'+time
            else:
                time= None

            re_penalty_on=re.search(r"PENALTY on ([A-Z]+).*, (.+), ([0-9]+).+,",str(mp))  
            if re_penalty_on:
                penalty_on=re_penalty_on.group(1)
            else:
                penalty_on= None

            re_penalty=re.search(r"PENALTY on ([A-Z]+).*, (.+), ([0-9]+).+,",str(mp))
            if re_penalty:
                penalty=re_penalty.group(2)
            else:
                penalty= None

            re_penalty_yards=re.search(r"PENALTY on ([A-Z]+).*, (.+), ([0-9]+).+,",str(mp))
            if re_penalty_yards:
                penalty_yards=re_penalty_yards.group(3)
            else:
                penalty_yards = None

            re_to_go=re.search(r"^\((\D+), (\D+\d+), (..), (\d+ and \d+)\) ",str(mp))
            if re_to_go:
                to_go=re_to_go.group(4)
            else:
                to_go= None
                
            if field_yards != 50:
                if team_possesion == field_team:
                    yards_to_td =  (50-field_yards)+50
                else:
                    yards_to_td=field_yards
            else:
                yards_to_td = field_yards
                
            #get winner and loser
            
            if ( game.score_home > game.score_away ):
                winner = game.schedule['home']
                loser = game.schedule['away']
            elif ( game.score_home == game.score_away ):
                winner = None
                loser = None
            else:
                loser = game.schedule['home']
                winner = game.schedule['away']
                            
        
            #Append vars to list
            list_play.append(year)
            list_play.append(game.schedule['home'])
            list_play.append(game.schedule['away'])
            list_play.append(game.schedule['week'])
            list_play.append(game.score_home)
            list_play.append(game.score_away)
            list_play.append(winner)
            list_play.append(loser)
            list_play.append(team_possesion)
            list_play.append(field_position)
            list_play.append(field_team)
            list_play.append(field_yards)
            list_play.append(yards_to_td)            
            list_play.append(quarter)
            list_play.append(to_go)
            list_play.append(time)
            list_play.append(penalty_on)
            list_play.append(penalty)
            list_play.append(penalty_yards)
            list_play.append(str(mp))

            #print(list_play)

            #Append to data list, the future df
            list_game_plays.append(list_play)
        
    
    return list_game_plays
    
    
```

```python
#main part


#Get info from regular season
#create the empty dataframe for all years
df_alltime = pd.DataFrame(columns = ['season','home','away','week','score_home','score_away','winner','loser','team_possesion',\
                                        'field_position','field_team','field_yards','yards_to_td','quarter','to_go','time_left_qtr','penalty_on',\
                                        'penalty','penalty_yards','full_description'])

for year in years:
    print(year)
    #print(weeks)
    try:
        year_games = nflgame.games(year, week = weeks , home='NE', away='NE')

        for game in year_games:
            #print (game.time)
            if str(game.time) != 'Pregame':  
                game_play_list= create_list_game_plays(game)

                df_game= pd.DataFrame(game_play_list, columns = ['season','home','away','week','score_home','score_away','winner','loser','team_possesion',\
                                        'field_position','field_team','field_yards','yards_to_td','quarter','to_go','time_left_qtr','penalty_on',\
                                        'penalty','penalty_yards','full_description'])

                df_alltime = pd.concat([df_alltime,df_game])
    except:
        print("Error al llamar nflgame.game con:" + year + week)

#add column for regular campaign flag
df_alltime['game_type']='REG'
```

```python
#Get info from playoffs
#create the empty dataframe for all years
df_alltime_playoffs = pd.DataFrame(columns = ['season','home','away','week','score_home','score_away','winner','loser','team_possesion',\
                                        'field_position','field_team','field_yards','yards_to_td','quarter','to_go','time_left_qtr','penalty_on',\
                                        'penalty','penalty_yards','full_description'])
for year in years:
    #print(year)
    #print(weeks)
    try:
        year_games_playoffs = nflgame.games(year, home='NE', away='NE', kind='POST')

        for game in year_games_playoffs:
            #print (game.time)
            if str(game.time) != 'Pregame':  
                game_play_list= create_list_game_plays(game)

                df_game= pd.DataFrame(game_play_list, columns = ['season','home','away','week','score_home','score_away','winner','loser','team_possesion',\
                                        'field_position','field_team','field_yards','yards_to_td','quarter','to_go','time_left_qtr','penalty_on',\
                                        'penalty','penalty_yards','full_description'])

                df_alltime_playoffs = pd.concat([df_alltime_playoffs,df_game])
    except:
        print("Error al llamar nflgame.game con:" + str(year))
        pass

#add column for regular campaign flag
df_alltime_playoffs['game_type']='POST'
```

```python
#concatenate regular and playoffs dataframe
df_alltime = pd.concat([df_alltime,df_alltime_playoffs])
```

```python
df_alltime['team_possesion_cat'] = df_alltime['team_possesion'].apply(lambda x: 'Opponent' if x!= 'NE' else x )
df_alltime['team_penalty_cat'] = df_alltime['penalty_on'].apply(lambda x: 'Opponent' if x != 'NE' else x)

df_alltime[:3]
```

```python
def get_penalty_side (dfr):
    if dfr['team_possesion_cat'] == dfr['team_penalty_cat']:
        return 'Offensive Penalty'
    else:
        return 'Defensive Penalty'
    
df_alltime['penalty_side']=df_alltime.apply(get_penalty_side,axis=1)
```

```python
# regex to opportunity an yards to go
def get_opportunity (to_go):
    re_opportunity= re.search(r"(\d+) and (\d+)",str(to_go))
    if re_opportunity:
        opportunity = re_opportunity.group(1)
        return opportunity
    else:
        return None

def get_yards_to_go (to_go):
    re_yards_to_go= re.search(r"(\d+) and (\d+)",str(to_go))
    if re_yards_to_go:
        yards_to_go = re_yards_to_go.group(2)
        return yards_to_go
    else:
        return None

    


df_alltime['opportunity']=df_alltime['to_go'].apply(get_opportunity)
df_alltime['yards_to_go']=df_alltime['to_go'].apply(get_yards_to_go)
#penalty_yards=re_penalty_yards.group(3)
```

```python
df_alltime.to_csv("Pats_penalty_plays_2009_2019.csv", index=False)
```
