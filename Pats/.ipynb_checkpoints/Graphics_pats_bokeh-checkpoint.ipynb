```python
import pandas as pd
```

```python
from bokeh.io import show, output_file,output_notebook
from bokeh.models import ColumnDataSource,FactorRange
from bokeh.plotting import figure
output_notebook()
```

```python
data = pd.read_csv('/Users/pablo.vazquez/Dropbox/PS/DataVizExercises/Pats/Pats_penalty_plays_2009_2019.csv')
data['Season']=data['Season'].astype(str)
data['team_penalty']= data['Penalty_on'].apply(lambda x: 'NE' if x =='NE' else 'Opponent')
data.head()
```

```python
#Create ColumnDataSource
source = ColumnDataSource(data)
```

```python
source.column_names
```

```python
x_range = FactorRange(factors=list(data.team_penalty))
```

```python
list(set(data.team_penalty))
```

```python
# create a Figure object
p = figure(
           x_range = FactorRange(factors=list(set(data.team_penalty))), #eliminate duplicates and use factor for categorical variables
           plot_width=300,
           plot_height=300,
           tools="pan,reset,save")
```

```python
p.vbar('team_penalty',top='Penalty_Yards',width=0.9,source=source)
show(p)
```

```python
#Make other approach to get the sum

grouped = pd.DataFrame(data.groupby(['Season','team_penalty'])['Penalty_Yards'].sum()).reset_index()
type(grouped)
#grouped
```

```python
source_g_season_team = ColumnDataSource(grouped)
```

```python
# create a Figure object
p_g = figure(
           x_range = FactorRange(factors=list(set(grouped.team_penalty))), #eliminate duplicates and use factor for categorical variables
           plot_width=300,
           plot_height=300,
           tools="pan,reset,save")

p_g.vbar('team_penalty',top='Penalty_Yards',width=0.9,source=source_g_season_team)
show(p_g)
```
